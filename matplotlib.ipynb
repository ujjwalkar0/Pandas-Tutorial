# Load Necessary Library

```python
import matplotlib.pyplot as pit
import numpy as np
import pandas as pd
```

## Basic Graphs

```python
pit.plot([1,2,3],[4,5,6])
```

```python
x=[1,2,3]
y=[4,5,6]

pit.plot(x,y,label='2X',color='green',linestyle='--',linewidth=2,marker='.',markersize=10,markeredgecolor='blue')
pit.title("My first graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
pit.xlabel("X axis")
pit.ylabel("Y axis")

# pit.xticks([0,1,2,3,4,])
# pit.yticks([0,2,4,6,8,10])
pit.legend()
pit.show()
```

```python
pit.plot(x,y,'r.-',label='2x')
pit.title("My first graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
pit.xlabel("X axis")
pit.ylabel("Y axis")

# pit.xticks([0,1,2,3,4,])
# pit.yticks([0,2,4,6,8,10])
pit.legend()
pit.show()
```

# Bar Chat

```python
labels = ['A','B','C']
values = [11,2,3]
bars=pit.bar(labels,values,color='green')
bars[0].set_hatch('o')
bars[1].set_hatch('*')
bars[2].set_hatch('/')

pit.figure(figsize=(20,20),dpi=100)
pit.show()
```
